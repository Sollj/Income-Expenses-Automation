## Creating our column categories, and Categorizing our Expenses for each transaction

```python
print("Current Working Directory ", os.getcwd())
os.chdir("C:/Users/james/Python Scripts/Personal Finance/Bank Transactions")
df = pd.read_csv("Bank Transactions Example.csv")

# Changing our Date column from Object type to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Changing categories for our df titles
##df.rename(columns = {'LEN':'Category'}, inplace=True) 

#Adding a new column (Category) to our DataFrame
df['Category'] = 'Category'
Category = np.arange(0,1000)

#Adding a new column (Account) to our DataFrame
df['Account'] = 'Account'
Account = np.arange(0,1000)

# Setting the max display of rows
pd.set_option('display.max_rows', 1000)

#Setting "Account" and "Category" column to object (str) type
df['Category'] = df['Category'].astype(str)
df['Account'] = df['Account'].astype(str)

##After running these lines of code, run the line of code below to change the data type to float for 
## the "Amount" column
df['Amount'] = df['Amount'].astype(str)

### CATEGORIZING OUR ACCOUNT TYPES FOR CATEGORIES
##LOOK TO CHANGE "Amount" COLUMN TO FLOAT TYPE AND STILL BE ABLE TO RUN THESE LINES OF CODE
df['Account'] = np.where(df['Amount'].str.contains('-')==False,'Revenue', df['Account'] )
df['Account'] = np.where(df['Amount'].str.contains('-')==True,'Expense', df['Account'] )

### These lines of code allows us change the category type to the desired category 
### given the string types in contains.
##########General Categories
#Other
#Food
df['Category'] = np.where(df['Description'].str.contains('Food'),'Food', df['Category'] )
#Subscriptions
df['Category'] = np.where(df['Description'].str.contains('Subscriptions'),'Subscriptions', df['Category'] )
#Car Insurance
df['Category'] = np.where(df['Description'].str.contains('Car Insurance'),'Car Insurance', df['Category'] )
#Parking
#Rent
df['Category'] = np.where(df['Amount'].str.contains('-250'),'Rent', df['Category'] )
#laundry
df['Category'] = np.where(df['Description'].str.contains('WASH LAUNDRY'),'Laundry', df['Category'] )
#Health Insurance
#Haircut
df['Category'] = np.where(df['Description'].str.contains('BISHOPS'),'Haircut', df['Category'] )

#Tax Payment
#Clothing
#Gym Membership

## Major purchases separately categorized

##Income
df['Category'] = np.where(df['Amount'].str.contains('-')==False,'Income', df['Description'])
```

# DEC 13th - JAN 13th: EXPENSES

```python
# Slices our dataframe into a monthly section of transactions, then deletes our row of data for
# income (row 402) 
df_total_exp_jan = df[(df['Date'] > '2021-12-13') & (df['Date'] < '2022-01-14')]
#df_total_exp_jan = df_total_exp_jan.drop(df.index[[444]])

# Switching the "Amount" column to a FLOAT type for calculations
df_total_exp_jan['Amount'] = df_total_exp_jan['Amount'].astype(float)

## Groups our January DataFrame by Category and sums each Category for January 
## (no cumsum of all categories)....
df_total_exp_jan = df_total_exp_jan.groupby(['Category']).sum()

## Groups our monthly DataFrame by Category and sums each Category for January  and resets the
## index name to sum, instead of amount
df_total_exp_jan.groupby('Category')['Amount'].sum().reset_index(name ='Sum')
```

```python
# .cumsum() our whole monthly dataframe to show us the total expense amount for the month
df_total_exp_jan['Amount'].sum()
```
